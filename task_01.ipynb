```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_130083.csv', skiprows=4)

plot_df=df[['Country Name', '2024']].sort_values('2024', ascending=False).head(10)
print(plot_df)
```

```python
plt.figure(figsize=(10,6))
plt.bar(plot_df['Country Name'], plot_df['2024'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Population in 2024')
plt.xlabel('Country')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```
